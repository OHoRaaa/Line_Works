```python
df_2 = pd.read_csv('C:\Big Data\라인웍스 자료(cdm_data)\\drug_exposure.csv')
df_2
```

```python
# 의약품 처방 데이터
df_2 = df_2[['person_id', 'drug_concept_id', 'drug_exposure_start_date', 'drug_exposure_end_date', 'visit_occurrence_id']]
df_2.head(5)
```

```python
df_3 = pd.read_csv('C:\Big Data\라인웍스 자료(cdm_data)\\condition_occurrence.csv')
df_3
```

```python
# 진단 데이터

df_3 = df_3[['person_id', 'condition_start_date', 'condition_end_date', 'visit_occurrence_id']]
df_3
```

```python
# merge
```

```python
tt = pd.merge(hos_i, df_2, on = ['visit_occurrence_id', 'person_id'], how = 'inner')
tt
```

```python
vcd = pd.merge(tt, df_3, on = ['visit_occurrence_id', 'person_id'], how = 'inner')
vcd
```

```python
# 칼럼명 편하게

vcd.columns = ['방문 ID', '환자 ID', '입원', '방문시작일', '방문종료일', '처방의약품', '처방시작일', '처방종료일', '진단시작일', '진단종료일']
vcd
```

```python
# 날짜칼럼을 datetime 형식으로 변환

vcd['진단종료일'] = pd.to_datetime(vcd['진단종료일'])
vcd['진단시작일'] = pd.to_datetime(vcd['진단시작일'])
vcd['처방시작일'] = pd.to_datetime(vcd['처방시작일'])
vcd['처방종료일'] = pd.to_datetime(vcd['처방종료일'])
```
